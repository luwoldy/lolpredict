```python
import pandas as pd
import requests
import json
import numpy as np
api_key = "RGAPI-562d0771-48bf-4db6-b87a-14b26f6d6ddc"
```

```python
# To get matches we need to get the summoner ID for a player, using the userinput

userinput = "MoreFiero"

def get_accid(summoner):
    url = f"https://na1.api.riotgames.com/lol/summoner/v4/summoners/by-name/{summoner}?api_key={api_key}"
    search = requests.get(url).json()
    summonid = search["accountId"]
    return summonid

acct_id = get_accid(userinput)
acct_id
```

```python
#Get matches based on the summoner id from get_accid function
match_ids = []
match_url = f"https://na1.api.riotgames.com/lol/match/v4/matchlists/by-account/{acct_id}?queue=420&api_key={api_key}"
results = requests.get(match_url).json()

for result in results["matches"]:
    match_ids.append(result["gameId"])
    
match_results = []    
for game in match_ids:
    match_id = f"https://na1.api.riotgames.com/lol/match/v4/matches/{game}?api_key={api_key}"
    match_result = requests.get(match_id).json()
    match_results.append(match_result)

# match_result
match_ids
```

```python
# run through last ten games 
match_results
```

```python
kills = []
deaths = []
longestTimeSpentLiving = []
wardsPlaced = []
goldEarned = []
champLevel = []
firstTowerAssist = []
firstInhibitorAssist = []
assists = []
firstBloodAssist = []
visionScore = []

count = 0

for match in match_results:
    count += 1
    print(f"Match #:{count}")
    print(f"Match #: {match['gameId']}")
    participant_list = match['participantIdentities']
    
    participant_info = []

    for participant in participant_list:
#         print(participant)
        participant_info.append(participant)

    players = []

    for player in participant_info:
#         print(player)
        players.append(player['player'])
    
    summoner_list = []
    summonercount = 0
    
    for summoner in players:
        print(summoner['summonerName'])
        summonercount += 1
        print(summonercount)
        summoner_list.append(summoner['summonerName'])
    
    participantNo = summoners.index(userinput)+1
    print(f"Participant No recorded = {participantNo}")

    participants = match['participants']
    
    statlist = []
    for k in participants:
#         print(k)
        statlist.append(k['stats'])

    for i in statlist:
        if participantNo == i['participantId']:
            kills.append(i['kills'])
            deaths.append(i['deaths'])
            longestTimeSpentLiving.append(i['longestTimeSpentLiving'])
            print(i['longestTimeSpentLiving'])
            wardsPlaced.append(i['wardsPlaced'])
            goldEarned.append(i['goldEarned'])
            champLevel.append(i['champLevel'])
            if i['firstTowerAssist']==True:
                firstTowerAssist.append(1)
            else: 
                firstTowerAssist.append(0)
            if i['firstInhibitorAssist']==True:
                firstInhibitorAssist.append(1)
            else: 
                firstInhibitorAssist.append(0)
            assists.append(i['assists'])
            if i['firstBloodAssist']==True:
                firstBloodAssist.append(1)
            else:
                firstBloodAssist.append(0)
            visionScore.append(i['visionScore'])
```
